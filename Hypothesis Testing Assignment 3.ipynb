# Tele Call Center
* Null Hypothesis: The defective percentage doesn't vary by center
* Alternative Hypothesis: The defective percentage varies by center

```python
import pandas as pd
import numpy as np
import scipy
from scipy import stats
```

```python
data = pd.read_csv('C:/Users/Dell/Desktop/Assignments/Hypothesis Testing/CostomerOrderForm.csv')
data.head()
```

```python
data1 = pd.get_dummies(data)
```

```python
data1
```

# Using Chi-square Method 

```python
result = scipy.stats.chi2_contingency(data1)
```

```python
result1 = [[' ', 'Test Statistic', 'p-value'], ['Sample Data', result[0], result[1]]]
result1
```

# Since the p-value > (0.05), we accept the Null Hypothesis
